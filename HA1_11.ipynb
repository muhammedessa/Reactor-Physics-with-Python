To solve the problem outlined in Home Assignment 1, we need to derive the equation that describes the evolution of uranium enrichment over time and implement a function to evaluate the enrichment at specific times in the past. We will then use this function to calculate the enrichment at the birth of Earth and when the Oklo reactor operated, and plot the enrichment over time.

Implement the Function
We will implement a function to evaluate the enrichment at a given time
t
t.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
half_life_U235 = 7.13e8  # years
half_life_U238 = 4.51e9  # years
lambda_U235 = np.log(2) / half_life_U235
lambda_U238 = np.log(2) / half_life_U238
epsilon_0 = 0.0072  # Initial enrichment (0.72%)

def epsilon(t):
    """Function to evaluate Uranium enrichment at time t."""
    ratio = (1 - epsilon_0) / epsilon_0
    exponent = np.exp(-(lambda_U238 - lambda_U235) * t)
    return 1 / (1 + ratio * exponent)
```

```python
# Times in years ago
time_birth_of_earth = 4.5e9
time_oklo_reactor = 1.7e9

# Calculate enrichments
enrichment_birth_of_earth = epsilon(time_birth_of_earth)
enrichment_oklo_reactor = epsilon(time_oklo_reactor)

print(f"Enrichment at the birth of Earth: {enrichment_birth_of_earth:.4f}")
print(f"Enrichment at the time of the Oklo reactor: {enrichment_oklo_reactor:.4f}")
```

Plot Enrichment vs Time
We will plot the enrichment over time and include vertical lines at the specified times

```python
# Generate a range of times
times = np.linspace(0, 5e9, 500)
enrichments = epsilon(times)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(times, enrichments, label='Uranium Enrichment')
plt.axvline(time_birth_of_earth, color='r', linestyle='--', label='Birth of Earth')
plt.axvline(time_oklo_reactor, color='g', linestyle='--', label='Oklo Reactor')
plt.xlabel('Time (years ago)')
plt.ylabel('Enrichment')
plt.title('Uranium Enrichment Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
